```python
import pandas as pd
import openpyxl

produtos_df = pd.read_excel("Produtos.xlsx")
display(produtos_df)

```

```python
## Atualizar o multiplicador
# Fazer a contra do preco base

produtos_df.loc[produtos_df["Tipo"] == "Serviço", ["Multiplicador Imposto"]] = 1.5

produtos_df["Preço Base Reais"] = produtos_df["Multiplicador Imposto"] * produtos_df["Preço Base Original"]
produtos_df.to_excel("Produtos_atualizado.xlsx", index=False)
```

```python
## openpyxl

from openpyxl import Workbook, load_workbook

planilha = load_workbook("Produtos_atualizado.xlsx")
aba_ativa = planilha.active

for celula in aba_ativa["C"]:
   if celula.value == "Serviço":
       linha = celula.row
       aba_ativa[f"D{linha}"] = 1.5
planilha.save("Produtos_atualizado_openpyxl.xlsx")       
```
